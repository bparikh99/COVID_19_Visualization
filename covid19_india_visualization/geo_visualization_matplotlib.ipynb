# Visualization by geopandas

### First import all necessary library

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Used shape file for visualization

```python
map_data=gpd.read_file("C:/Users/himanshimehta/Downloads/Igismap/Indian_States.shp")
```

### After getting dataframe  replace state_name 

```python
map_data['st_nm'] = map_data['st_nm'].str.replace('&', 'and')
map_data['st_nm'].replace('Arunanchal Pradesh', 'Arunachal Pradesh', inplace = True)
map_data['st_nm'].replace('Telangana', 'Telengana', inplace = True)
map_data['st_nm'].replace('NCT of Delhi', 'Delhi', inplace = True)
map_data['st_nm'].replace("Andaman and Nicobar Island",'Andaman and Nicobar Islands',inplace=True)
```

### Get the data and clean it!!

```python
dfs=pd.read_html("https://www.mohfw.gov.in/")
df=dfs[0]
df.columns
df.drop(columns=['S. No.'],inplace=True,axis=1)
df=df.rename(columns={'Name of State / UT':'st_nm',
       'Total Confirmed cases (Including 111 foreign Nationals)':"confirmed",
       'Cured/Discharged/Migrated':"deasesed",
       'Deaths ( more than 70% cases due to comorbidities )':"deaths"})
df.drop(index=[32,33,34],inplace=True)
```

```python
df
```

### plot of map_data

```python
map_data.plot()
```

### Merged DataFrame

```python
merged_data=pd.merge(map_data,df,how="left",on='st_nm')
merged_data
```

### Visualization Part

```python

merged_data.fillna(0, inplace = True)
merged_data['confirmed']=merged_data['confirmed'].astype(int)

# create figure and axes for Matplotlib and set the title
fig, ax = plt.subplots(1, figsize=(20, 12))
ax.axis('off')
ax.set_title('Covid-19 Statewise Data - Confirmed Cases', fontdict = {'fontsize': '25', 'fontweight' : '3'})
# plot the figure
merged_data.plot(column = 'confirmed', cmap='Reds', linewidth=0.8, ax=ax, edgecolor='0.8', legend = True)
plt.show()
```
